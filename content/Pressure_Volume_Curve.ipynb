### Basic Pressure-Volume Curve
This is the most basic mathematical underpinning of a gravity balloon.

old (largely deprecated) post: https://gravitationalballoon.blogspot.com/2013/03/gravity-balloon-pressure-volume-curve.html

updated derivations: https://gravitationalballoon.blogspot.com/2021/07/a-more-detailed-run-through-of-pressure.html

The first graphs produced here (P-R and P-V curves) are constant-mass curves. You could mentally consider this an "inflation" process, where an inner cavity increases in size as air is added. In practice, a gravity balloon is likely to require a static liner, so this is unlikely to be realistic. It is purely a mental exercise to understand the mathematics.

```python
from gb.inflation import P_RM, P_VM
from gb.constants import rho

from math import pi

import matplotlib.pyplot as plt

from IPython.display import Math
```

General setup, pick arbitrary mass, define constants, make a scaffolding for the x-axis

```python
M = 1.e14
N_points = 200

R0 = (M * 3. / (4. * pi * rho))**(1./3.)
V0 = (4./3.)*pi*R0**3

relative_values = [(2. / N_points) * i for i in range(N_points + 1)]
```

Here, I will echo the equations being solved.

```python
Math(
    r'P = \frac{2}{3} G \pi \rho^2 t^2 \left( \frac{3 R + t}{R + t} \right) \\'
    r'M = \frac{4}{3} \pi \rho \left( \left( R + t \right)^3 - R^3 \right)'
)
```

Create plot of the relative Pressure-Radius curve

```python
R_values = [rel * R0 for rel in relative_values]
P_values = [P_RM(R, M, rho=1.3e3) / P_RM(0., M, rho=1.3e3) for R in R_values]

plt.plot(relative_values, P_values)
plt.title('Pressure vs. Inner Radius, constant Mass')
plt.ylabel('P relative to original core P')
plt.xlabel('Radius relative to asteroid radius')
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.axis([0., 2., 0., 1.])
plt.show()
```

```python
# Important that we clear the plot before doing the next one
plt.clf()
```

Create plot of the relative Pressure-Volume curve

```python
V_values = [rel * V0 for rel in relative_values]
PV_values = [P_VM(V, M) / P_VM(0., M) for V in V_values]

plt.plot(relative_values, PV_values)
plt.title('Volume vs. Inner Radius, constant Mass')
plt.ylabel('P relative to original core P')
plt.xlabel('Volume relative to asteroid rock volume')
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.axis([0., 2., 0., 1.])
plt.show()
```
